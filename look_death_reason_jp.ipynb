## 総数、男、女に分解

```python
dr = dr_org.copy()
t_index = 0
m_index = dr.loc[dr[dr.columns[0]]=="男"].index[0]
f_index = dr.loc[dr[dr.columns[0]]=="女"].index[0]
```

```python
dr.loc[t_index:m_index-1,("gender_0","gender")] = "総数"
dr.loc[m_index+1:f_index-1,("gender_0","gender")] = "男"
dr.loc[f_index+1:,("gender_0","gender")] = "女"
```

```python
dr.groupby(("gender_0","gender")).count()
```

```python
dr[dr[("gender_0","gender")].isna()]
```

```python
dr.dropna().loc[dr[("gender_0","gender")].isna()]
```

```python
dr.dropna(inplace=True)
```

```python
dr[dr[("gender_0","gender")].isna()]
```

## 縦横の入れ替え

```python
# キーのデータ切り出し
dr_k = dr[[dr.columns[0],dr.columns[-1]]]
dr_k.columns = dr_k.columns.droplevel()
dr_k = dr_k.rename(columns={dr_k.columns[0]:"year"})
```

```python
# 数のデータ縦横変換
dr_d = dr.iloc[:,1:-1].stack(level=0).drop(labels="死亡率",axis=1).dropna().rename(index={"Unnamed: 1_level_0":"総数"})
dr_d = dr_d.reset_index().rename(columns={"level_1":"死因"})
```

```python
# キーとデータ結合
drs = pd.merge(dr_k,dr_d,left_index=True,right_on="level_0", how="left").drop(columns="level_0")
drs
```
